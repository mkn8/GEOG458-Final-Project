# Scraping the locations of the article: "Following The Money Trail To Alaska's 'Ballard North'"

```python
import cssselect
import urllib2
import lxml
from lxml import html
import ssl
```

```python
import folium
import geopandas
import xml.etree.ElementTree as ET 
import pandas
import urllib
import json
```

```python
#THIS CODE BELOW IS FOR: http://kuow.org/post/following-money-trail-alaskas-ballard-north
#here is where we request access to the text in the link 
url = "http://kuow.org/post/following-money-trail-alaskas-ballard-north"
req = urllib2.Request(url, headers={'User-Agent' : "Magic Browser"}) 
con = urllib2.urlopen( req )
doc_text = con.read()
doc = lxml.html.fromstring(doc_text)
doc.make_links_absolute(url)
```

```python
#this block extracts the article's text, which will be used in geoparser.io 
story = ""
for row in doc.cssselect("div.field-item.even"):
    whole_body = row.cssselect("p")
    for i in whole_body:
        temp = i.text

        if temp != None:
            temp = temp.strip()
            if (temp.startswith("In Ballard")) or (temp.startswith("SEAN CASADY")) or (temp.startswith("The yard")) or (temp.startswith("On Monday")) or (temp.startswith("Sonny")) or (temp.startswith("KIRO:")):
                #removed bottom stories from the webpage because that info was not needed 
                print ""
            else:

                story = story + " " + temp
        
print story
```

# The code below is specifically for the first story of our narrative map: 

```python
#for i in geo_story: #this shows that geo_story (the output from geoparser.io) is a string of geojson
#    print i
```

```python
json_as_dictionary = json.loads(geo_story) #this puts all of the geojson data into a dictionary
```

```python
json_as_dictionary #this shows what is currently in the dictionary we just created
```

```python
json_as_dictionary['features'] #this looks into the first key of the dictionary because there are multiple keys
```

```python
tableit = []

#note that the coordinates are listed as [longitude, latitude]

for x in json_as_dictionary['features']: #this block of code goes into the dictionary of dictionaries to specifically pull out coordinates
    test = x['geometry']['coordinates']
    longit = test[0]
    latit = test[1]
    
    dict = {'Location':x['properties']['name'], 'Lat':latit, 'Long':longit}
    tableit.append(dict)
 
df = pandas.DataFrame(tableit, columns=['Location','Lat', 'Long'])
df
```

```python
df.head(3) #tests to see what the first 3 rows of the DF look like
```

```python
import shapely
import shapely.geometry
import fiona
import fiona.crs
```

```python
CoordinateTuples_list = zip(df['Long'], df['Lat']) #unites the lat and lon values into a tuple
CoordinateTuples_list
```

```python
geometry_list = [shapely.geometry.Point(CoordinateTuple) for CoordinateTuple in CoordinateTuples_list] #changes the coordinates into usable geometry points
geometry_list
```

```python
geometry_gs = geopandas.GeoSeries(geometry_list) #changes the coordinate tuple to a series 
geometry_gs
```

```python
%matplotlib inline

geopandas_gdf = geopandas.GeoDataFrame( #this adds the geometry column to the original table
    df,
    geometry=geometry_gs,
)
geopandas_gdf.crs=fiona.crs.from_epsg(4326) #this sets the projection of the map

geopandas_gdf #this is now what the table looks like with the geometry field added so that our data is mappable
```

```python
geopandas_gdf = geopandas_gdf.drop(0) #dropped the row containing a location we do not need in our narrative map - Mt. Seattle 
geopandas_gdf = geopandas_gdf.drop(1) # Anchorage
geopandas_gdf = geopandas_gdf.drop(4) # N. America
geopandas_gdf = geopandas_gdf.drop(5) # US Bend
geopandas_gdf = geopandas_gdf.drop(8) # Turabah
geopandas_gdf = geopandas_gdf.drop(10) # Alaska
geopandas_gdf = geopandas_gdf.drop(11) # US

geopandas_gdf.to_file('map_article1.geojson', driver = 'GeoJSON') #this creates a geojson file for the first article used for the narrative map
```

```python
geopandas_gdf #this is the final table we used and the coordinates we used to create our narrative map 
```
