#### Import pandas

#### Make dataframes of csv files

```python
boroughs_orig_df = pd.read_csv("Borough_Coords_Boroughs2016Immigration.csv")
countries_orig_df = pd.read_csv("Country_Coords_Boroughs2016Immigration.csv")
```

#### Get lists of countries and boroughs from csv files

```python
boroughs_csv = open("Borough_Coords_Boroughs2016Immigration.csv")
boroughs = []

for line in boroughs_csv:
    borough = line.split(",")[0] + line.split(",")[1]
    boroughs.append(borough)
        
boroughs.remove('Boroughs_and_RegionsTotal')
#print boroughs
#print len(boroughs)
```

```python
countries_csv = open("Country_Coords_Boroughs2016Immigration.csv")
countries = []

for line in countries_csv:
    country = line.split(",")[0]
    countries.append(country)
        
countries.pop(0)
#print countries
#print len(countries)
```

#### Import Nominatim

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

#### Using Nominatim, search for (and append to the list, "data") the coordinates and related information corresponding to each borough and country

```python
boroughs_data = []
for borough in boroughs:
    input_address = borough
    if geolocator.geocode(input_address):
        location = geolocator.geocode(input_address)
        boroughs_data.append(location.raw)
#print boroughs_id
```

```python
countries_data = []
for country in countries:
    input_address = country
    if geolocator.geocode(input_address):
        location = geolocator.geocode(input_address)
        countries_data.append(location.raw)
#print countries_id
```

#### Clean the "lon", "lat", and "boundingbox" keys in each dictionary in the list, "data" so they can be converted to a geodataframe later on

```python
for i in range(len(countries_data)):
    countries_data[i]['lon'] = float(countries_data[i]['lon'])
    countries_data[i]['lat'] = float(countries_data[i]['lat'])
    countries_data[i]['boundingbox'] = ''.join(countries_data[i]['boundingbox'])

for i in range(len(boroughs_data)):
    boroughs_data[i]['lon'] = float(boroughs_data[i]['lon'])
    boroughs_data[i]['lat'] = float(boroughs_data[i]['lat'])
    boroughs_data[i]['boundingbox'] = ''.join(boroughs_data[i]['boundingbox'])
    
```

#### Convert boroughs_data and countries_data into dataframes

```python
boroughs_df = pd.DataFrame(boroughs_data)
countries_df = pd.DataFrame(countries_data)
```

#### Join the dataframes that were just made with the dataframes made at the beginning of the program

```python
join_boroughs_df = pd.concat([boroughs_df, boroughs_orig_df], axis=1)
join_boroughs_df
```

```python
join_countries_df = pd.concat([countries_df, countries_orig_df], axis=1)
join_countries_df
```

#### Import shapely, shapely.geometry, fiona, fiona.crs, and geopandas

```python
import shapely
import shapely.geometry
import fiona
import fiona.crs
import geopandas
```

#### Convert joined dataframes into geodataframes

```python
#print
CoordinateTuples_list = zip(join_boroughs_df['lon'], join_boroughs_df['lat'])
#print "CoordinateTuples_list: "
#print CoordinateTuples_list 

#print
geometry_list = [shapely.geometry.Point(CoordinateTuple) for CoordinateTuple in CoordinateTuples_list]
#print "geometry_list: "
#print geometry_list

#print
geometry_gs = geopandas.GeoSeries(geometry_list)
#print "geometry_gs, a GeoSeries:"
#print str(geometry_gs) 

#print
join_boroughs_gdf = geopandas.GeoDataFrame(join_boroughs_df, geometry=geometry_gs,)
join_boroughs_gdf.crs=fiona.crs.from_epsg(4326)   # don't miss this line.
#print "GeoDataFrame:"
join_boroughs_gdf.head()
```

```python
#print
CoordinateTuples_list = zip(join_countries_df['lon'], join_countries_df['lat'])
#print "CoordinateTuples_list: "
#print CoordinateTuples_list 

#print
geometry_list = [shapely.geometry.Point(CoordinateTuple) for CoordinateTuple in CoordinateTuples_list]
#print "geometry_list: "
#print geometry_list

#print
geometry_gs = geopandas.GeoSeries(geometry_list)
#print "geometry_gs, a GeoSeries:"
#print str(geometry_gs) 

#print
join_countries_gdf = geopandas.GeoDataFrame(join_countries_df, geometry=geometry_gs,)
join_countries_gdf.crs=fiona.crs.from_epsg(4326)   # don't miss this line.
#print "GeoDataFrame:"
join_countries_gdf.head()
```

#### Convert geodataframes into csv files; add csv files to folder
